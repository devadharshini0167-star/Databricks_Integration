```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Example: Find primes in range 1 to 100
primes = [x for x in range(1, 101) if is_prime(x)]
display(primes)
```
